# 7. Write a Python program to plot two or more lines with different styles

```python
import matplotlib.pyplot as plt
x1 = [10,20,30]
y1 = [20,40,10]
x2 = [10,20,30]
y2 = [40,10,30]
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.plot(x1,y1, color='black', linewidth = 2,  label = 'line1-dotted',linestyle='dotted')
plt.plot(x2,y2, color='yellow', linewidth = 3,  label = 'line2-dashdot', linestyle='dashdot')
plt.title("Plot with two or more lines with different styles")
plt.legend()
plt.show()
```
