# 8. Write a Python program to plot two or more lines and set the line markers. 

```python
import matplotlib.pyplot as plt
x = [1,4,5,6,7]
y = [2,6,3,6,3]

plt.plot(x, y, color='black', linestyle='dotted', linewidth = 2,
         marker='o', markerfacecolor='red', markersize=15)
plt.xlabel('x - axis')
plt.ylabel('y - axis')

plt.title('Display marker')
plt.show()
```
