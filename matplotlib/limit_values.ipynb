#  9. Write a Python program to display the current axis limits values and set new axis values. 

```python
import matplotlib.pyplot as plt
X = range(1, 50)
Y = [value * 3 for value in X]
plt.plot(X, Y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Draw a line.')
print("Current axis limit values: ", plt.axis()) 
plt.show()
plt.plot(X, Y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Extended axis.')
plt.axis([0, 100, 0, 200]) 
# Display the figure.
plt.show()
```
