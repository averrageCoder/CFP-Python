# 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

```python
import matplotlib.pyplot as plt
X = range(1, 50)
Y = [value * 2 for value in X]
print("Values of X:")
print(*X) 
print("Values of Y (twice of X):")
print(Y)
plt.plot(X, Y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Draw a line.')
plt.show()
```
