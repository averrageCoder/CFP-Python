```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from pylab import randn
import random
import plotly.graph_objects as go
```

```python
df = pd.read_csv('HR_Employee_Attrition.csv')
data_attr_yes = df[df.Attrition == 'Yes']
data_attr_no = df[df.Attrition == 'No']
df.head()
```

```python
df.describe()
```

# Organization Age Distribution

```python
plt.figure(figsize=(12, 6))
plt.title('Organization Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Employees')
bins = np.arange(20,70,5)
plt.hist(df.Age, bins=bins, label='Attrition', rwidth=0.8)

plt.xticks(bins)
plt.show()
```

### Observation:
- It can be found that the organisation has more employee in the age range 30-40

```python
bins = np.arange(20,70,5)
plt.title('Organization Attrition Age Distribution')
plt.hist([data_attr_yes.Age, data_attr_no.Age], bins=bins,stacked=True,  rwidth=0.8)
plt.xticks(bins)
plt.legend(['Yes', 'No'])
plt.show()
```

### Observation:
- It can be seen that nearly half of employee of age 20-25 attried

```python
bins = np.arange(20,70,5)
plt.figure(figsize=(12, 6))
plt.title('Organization Attriton Age Distribution')
plt.xlabel('Age')
plt.ylabel('Number of Employees left the organization')
plt.hist(data_attr_yes.Age, bins=bins, label='Attrition', rwidth=0.8)
plt.xticks(bins)
plt.show()
```

### Observation:
- From the above graph, it can be analyzed that more employee have attried b/w the age of 20-25

```python
cm = plt.cm.RdBu_r
bins = np.arange(0,30,3)
plt.figure(figsize=(12, 6))
plt.title('Organization Distance Distribution')
plt.xlabel('Distance in Miles')
plt.ylabel('Number of Employees left the organization')
# bins = np.arange(20,70,5)
n, bins, patches = plt.hist([data_attr_yes.DistanceFromHome], bins=bins,stacked=True, rwidth=0.8)
plt.xticks(bins)
# plt.legend(['Yes', 'No'])
for i, p in enumerate(patches):
    plt.setp(p, 'facecolor', cm(i/25)) # notice the i/25
plt.show()
```

### Observation:
- It can be found that employee with 0-3 miles of the office attried more

```python
sns.set(rc={'figure.figsize':(15,8)})
sns.countplot(x="YearsAtCompany", hue="Attrition", data=df);
```

### Observation:
- From the above graph, we see that employees that have worked in the company for 1-2 years attried more than others

```python
sns.set(rc={'figure.figsize':(10,6)})
sns.scatterplot(data=df, x="YearsAtCompany", y="MonthlyIncome", hue="Attrition")
```

### Observation:
- It can be seen from the above graph that there is a huge cluster of employee from 0-5 yrs at company and monthly income less than 7500.
- And from this huge cluster many employess have attried

```python
di = {1: "Below College", 2: "College", 3: "Bachelor", 4: "Master", 5: "Doctor"}
explode = (0.1, 0.05, 0.05, 0.05,0.05)  
education = data_attr_yes.replace({"Education": di})
education['Education'].value_counts().plot(kind='pie', explode=explode, autopct='%.1f%%')
```

### Observation:
- From the above graph, it can be found that employees with an education of Bachelors attried the most

```python
f, axs = plt.subplots(1, 2, figsize=(14,6))
sns.boxplot(y="JobSatisfaction", x="PerformanceRating", hue="Attrition", data=df, ax=axs[0])

sns.boxplot(y="WorkLifeBalance", x="PerformanceRating", hue="Attrition", data=df, ax=axs[1])
```

### Observation:
- From the above box plot, we can say that jobSatisfaction is more corelated to attrition as compared to WorkLifeBalance

```python
f, axs = plt.subplots(1, 2, figsize=(14,6))
sns.pointplot(y="JobSatisfaction", x="PerformanceRating", hue="Attrition", linestyles = '-.', data=df, 
            ax=axs[0])

sns.pointplot(y="JobSatisfaction", x="WorkLifeBalance", hue="Attrition", linestyles = '-.', data=df, 
            ax=axs[1])
```

### Observation:
- From the above graph, we see as performance increase, the job satisfaction remains less for employees who attried.
- Same thing can be seen for work life balance against job satisfaction too
